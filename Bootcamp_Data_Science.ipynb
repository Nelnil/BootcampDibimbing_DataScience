```python
# 1. Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load Dataset
df = pd.read_csv("/content/survey.csv")
print("Ukuran dataset:", df.shape)
df.head()
```

```python
# 3. Lihat Sekilas Data
print(df.head())
print(df.info())
print(df.describe(include='all'))
```

```python
# 4. Cek Missing Value
missing_values = df.isnull().sum()
print("Missing values per kolom:\n", missing_values[missing_values > 0])
```

```python
# Visualisasi missing value
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Heatmap Missing Values")
plt.show()
```

```python
# 5. Cek Duplicate
duplicates = df.duplicated().sum()
print(f"Jumlah baris duplikat: {duplicates}")
```

```python
# 6. Distribusi Respon (Contoh Kolom)
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="treatment", palette="Set2")
plt.title("Apakah Responden Pernah Mendapat Perawatan Kesehatan Mental?")
plt.xlabel("Pernah Mendapat Perawatan")
plt.ylabel("Jumlah Responden")
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Korelasi antar fitur numerik (jika ada)
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
if not numeric_cols.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
    plt.title("Korelasi Fitur Numerik")
    plt.show()
```
